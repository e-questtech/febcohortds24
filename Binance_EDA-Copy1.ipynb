```python
#Get sample of historical data
sampleh=client.get_historical_klines("ETHBRL", Client.KLINE_INTERVAL_5MINUTE, "2 Jan 2023")
df_sampleh = pd.DataFrame(sampleh)
df_sampleh.columns=["Open Time","Open","High","Low","Close","Volume","Close Time","Quote Asset Volume","Number of Trades","TB Base Volume","TB Quote volume","Ignore"]
df_sampleh["Open Time"]=pd.to_datetime(df_sampleh["Open Time"]/1000, unit="s")
df_sampleh["Close Time"]=pd.to_datetime(df_sampleh["Close Time"]/1000, unit="s")
num_columns=["Open","High","Low","Close","Volume","Quote Asset Volume","TB Base Volume","TB Quote volume"]
df_sampleh[num_columns]=df_sampleh[num_columns].apply(pd.to_numeric,axis=1)
```
